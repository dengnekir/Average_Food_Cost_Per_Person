# ***PDF ve CDF Hesabı***

### ***Veri Seçimi: Sütun olarak 2012-10-01 00:00:00 seçildi. Bu sütun, her bir eyalet için kişi başına düşen gıda maliyetlerini temsil eder.***

PDF Hesaplama: PDF (Probability Density Function), verilerin olasılık yoğunluğunun fonksiyonudur. Burada Kernel Density Estimation (KDE) kullanılarak veri dağılımı tahmin ediliyor.

CDF Hesaplama: CDF (Cumulative Distribution Function), verilerin kümülatif olasılıklarını hesaplar. PDF'nin birikimli toplamı olarak bulunur.

**Görselleştirme: Hem PDF hem de CDF'yi yan yana görselleştirmek için matplotlib kütüphanesi kullanıldı.**

*Olasılık Hesaplama: Belirli bir aralıktaki (örneğin, 30 ile 50 arasındaki değerler) olasılık, KDE'nin integrate_box() fonksiyonu kullanılarak hesaplandı.*

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# CSV dosyasını okuma
file_path = "Average_Food_Cost_Per_Person.csv"  # Dosya yolunu belirtin
df = pd.read_csv(file_path)

# Sayısal verilerin olduğu bir sütun seçelim (örneğin, 2012-10-01)
column = '2012-10-01 00:00:00'

# Seçilen sütunun boş olmayan değerlerini alıyoruz
data = df[column].dropna()

# PDF (Probability Density Function) hesaplama
# PDF için Kernel Density Estimation (KDE) kullanıyoruz
kde = stats.gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)
pdf = kde(x)

# CDF (Cumulative Distribution Function) hesaplama
cdf = np.cumsum(pdf) * (x[1] - x[0])

# Görselleştirme
plt.figure(figsize=(12, 6))

# PDF Grafiği
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title(f'{column} - PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# CDF Grafiği
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='red')
plt.title(f'{column} - CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()

# İsteğe bağlı olarak belirli bir veri aralığının olasılığını hesaplamak (örneğin, 30 ile 50 arasında)
lower_bound = 30
upper_bound = 50
probability = kde.integrate_box(lower_bound, upper_bound)

print(f"Olasılık (30 ile 50 arasındaki veri): {probability:.4f}")
```

# ***Cdf ve Pdf nin histogram olarak çizilmesi***

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# CSV dosyasını okuma
file_path = "Average_Food_Cost_Per_Person.csv"  # Dosya yolunu belirtin
df = pd.read_csv(file_path)

# Sayısal verilerin olduğu bir sütun seçelim (örneğin, 2012-10-01)
column = '2012-10-01 00:00:00'

# Seçilen sütunun boş olmayan değerlerini alıyoruz
data = df[column].dropna()

# PDF (Probability Density Function) hesaplama
# PDF için Kernel Density Estimation (KDE) kullanıyoruz
kde = stats.gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)
pdf = kde(x)

# CDF (Cumulative Distribution Function) hesaplama
cdf = np.cumsum(pdf) * (x[1] - x[0])

# Histogram (veri dağılımı)
plt.figure(figsize=(12, 6))

# Histogram ve PDF'yi aynı grafikte çizme
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, density=True, alpha=0.6, color='g', label='Histogram')
plt.plot(x, pdf, label='PDF', color='blue', linewidth=2)
plt.title(f'{column} - PDF ve Histogram')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Histogram ve CDF'yi aynı grafikte çizme
plt.subplot(1, 2, 2)
plt.hist(data, bins=20, density=True, alpha=0.6, color='g', label='Histogram')
plt.plot(x, cdf, label='CDF', color='red', linewidth=2)
plt.title(f'{column} - CDF ve Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()
```

# *** Veri Dağılımı Yorumu***
# 1. PDF ve Histogram (Sol Grafik)
*** Histogram: *** Histogram, veri setindeki her bir değerin sıklığını gösteren bir çubuk grafik türüdür. Burada histogram, belirli bir aralıktaki gıda maliyetlerinin sıklığını gösterir. Çubuğun yüksekliği, o aralıktaki veri noktalarının yoğunluğunu temsil eder.

***PDF (Probability Density Function):*** PDF, veri setindeki her bir değerin olasılık yoğunluğunu gösterir. Grafik üzerinde mavi çizgi ile gösterilen PDF, belirli bir değerin olasılık yoğunluğunun ne kadar olduğunu bize anlatır. Eğer bir bölge daha fazla yoğunluk gösteriyorsa, bu bölgeyi temsil eden değerlerin daha sık görüldüğünü ifade eder. PDF'nin yüksek olduğu bölgeler, bu değerlerin daha fazla sıklıkla görüldüğünü gösterir.

***Yorum***:

Histogramın ve PDF'nin şekli, verinin yoğunlaştığı aralıkları gösterir. Eğer PDF eğrisi yüksekse, bu aralıktaki verilerin daha sık olduğunu gösterir.
Verinin bir "tepe noktası" veya "mod" bulundurması, bu verinin en yaygın olduğu aralığı gösterir.
Verinin ne kadar yayılacağı, dağılımının genişliği ile ilgili fikir verir. Eğer PDF dar bir alanda yoğunlaşıyorsa, veriler belirli bir aralığa sıkışmıştır.
# 2. CDF ve Histogram (Sağ Grafik)
***CDF (Cumulative Distribution Function):*** CDF, verilerin kümülatif olasılıklarını gösterir. Yani, belirli bir değere kadar olan verilerin toplam olasılığını temsil eder. CDF eğrisinin yukarıya doğru çıkması, verilerin artan değerlerinin olasılıklarının birikmesini gösterir.
Yorum:

CDF eğrisinin dik bir şekilde yükseldiği bölgeler, verilerin hızlı bir şekilde biriktiği yerlerdir. Bu, verilerin o aralıkta yoğun olduğunu gösterir.
CDF'nin daha yatay olduğu bölgelerde, veriler daha geniş bir aralığa yayılmıştır ve kümülatif olasılık daha yavaş artar.
CDF, verilerin belirli bir değere kadar olan toplam olasılıklarını gösterdiği için, veri setindeki olasılıkların nasıl dağıldığını daha net bir şekilde anlamamızı sağlar.
# Genel Yorum:
***Veri Yoğunluğu:*** PDF ve CDF'yi incelediğimizde, verinin belirli bir aralıkta yoğunlaştığını görebiliriz. Eğer PDF eğrisi belirli bir değerin etrafında yüksekse, bu, gıda maliyetlerinin o aralıkta sıkça görüldüğünü ve bu değerlerin "standart" değerler olarak kabul edilebileceğini gösterir.

***Veri Dağılımı:*** Veri setinin belirli bir değere odaklanması veya geniş bir aralığa yayılması, veri dağılımı hakkında bilgi verir. Verinin çok yayılmadığı bir dağılım, verilerin daha tutarlı ve belli bir değere yakın olduğunu, geniş bir dağılım ise verilerin çok çeşitlendiğini gösterir.

Sonuç olarak, PDF ve CDF, veri dağılımını analiz etmek ve anlamak için güçlü araçlardır. Bu tür görselleştirmeler, verinin karakteristiklerini ve olasılıklarını anlamada bize yardımcı olur.

# **8.  Bar Chart (Çubuk Grafik)**

```
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
filtered_df = pd.read_csv('Average_Food_Cost_Per_Person.csv')

# 'State Agency or Indian Tribal Organization' ve 'Cumulative Average' hariç tutarak, her satırın toplam maliyetini hesapla
total_cost = filtered_df.drop(columns=['State Agency or Indian Tribal Organization', 'Cumulative Average']).sum(axis=1)
filtered_df['Total Cost'] = total_cost

# Eyaletlere göre toplam gıda maliyetini gruplama
total_cost_by_state = filtered_df.groupby('State Agency or Indian Tribal Organization')['Total Cost'].sum()

# Çubuk grafik: Eyaletlerin toplam gıda maliyeti
plt.figure(figsize=(12, 8))
total_cost_by_state.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title("Eyaletlerin Toplam Gıda Maliyeti")
plt.xlabel("Eyalet")
plt.ylabel("Toplam Gıda Maliyeti")
plt.xticks(rotation=90)
plt.show()
```

# 1. Scatter Plot (Dağılım Grafiği)
Scatter plot, iki değişken arasındaki ilişkiyi görselleştirmek için kullanılır. Bu grafik, her bir eyaletin belirli bir zaman dilimindeki gıda maliyeti ile diğer bir zaman dilimindeki maliyeti arasındaki ilişkiyi gösterebilir. Örneğin, Ekim 2012 ve Kasım 2012 arasındaki maliyetler arasındaki ilişkiyi inceleyebilirsiniz. Veri noktalarının nasıl kümelendiğini ve herhangi bir doğrusal ilişki olup olmadığını gözlemleyebilirsiniz.

# 2. Pie Chart (Pasta Grafiği)
Pasta grafiği, her bir eyaletin toplam gıda maliyetinin toplam maliyete oranını görsel olarak gösterir. Her dilim, eyaletlerin toplam maliyetten aldığı yüzdelik payı temsil eder. Bu tür bir görselleştirme, hangi eyaletlerin toplam maliyette daha büyük bir paya sahip olduğunu kolayca anlamanızı sağlar. Büyük dilimler daha yüksek maliyetleri, küçük dilimler ise daha düşük maliyetleri temsil eder.

# 3. Box Plot (Kutu Grafiği)
Box plot, veri setindeki dağılımı özetleyen bir grafik türüdür. Kutu grafiği, verinin medyanını, çeyrek değerlerini (Q1, Q3), ve uç değerlerini (outliers) gösterir. Bu grafiği kullanarak, belirli bir eyaletin maliyetlerinin ne kadar değişken olduğunu, ne kadar simetrik olduğunu ve uç noktaların olup olmadığını inceleyebilirsiniz. Eğer çok sayıda uç değer varsa, bu, gıda maliyetlerinin geniş bir aralıkta değiştiğini gösterir.

# 4. Histogram (Histogram Grafiği)
Histogram, veri dağılımını sınıflandırarak gösterir. Veri, belirli aralıklara (binler) bölünür ve her bir binin yüksekliği, o aralıktaki gözlem sayısını temsil eder. Gıda maliyetleri için histogram çizildiğinde, maliyetlerin hangi aralıklarda yoğunlaştığı ve ne kadar yayılım gösterdiği anlaşılabilir. Örneğin, çoğu eyaletin düşük maliyetlerde yoğunlaşması ya da belirli bir maliyet seviyesinin yoğunlaşması gibi desenler gözlemlenebilir.

# 5. Violin Plot (Viyolin Grafiği)
Viyolin grafiği, dağılımın yoğunluğunu ve şeklinin değişimini daha iyi görselleştiren bir grafik türüdür. Box plot'a benzer şekilde veri setindeki dağılımı gösterir, ancak aynı zamanda yoğunluğu daha ayrıntılı bir şekilde sergiler. Bu, verinin belirli bir aralıktaki yoğunluğunun ve dağılımının ne kadar simetrik veya asimetrik olduğunu gösterir.

# 6. KDE Plot (Kernel Density Estimate)
KDE, verinin sürekli dağılımını tahmin etmek için kullanılan bir tekniktir. Bir yoğunluk fonksiyonu üzerinden veri setinin olasılık dağılımını gösterir. KDE plot, verinin sürekli bir şekilde nasıl dağıldığını anlamamıza yardımcı olur. Gıda maliyetlerinin hangi aralıklarda daha yoğun olduğunu gösteren bir çizgi ile veri yoğunluğunun dağılımını gözlemleyebilirsiniz. Eğer eğri simetrikse, verinin normal dağıldığını gösterebilir.

# 7. Heatmap (Isı Haritası)
Isı haritası, genellikle korelasyon gibi sayısal veriler arasındaki ilişkileri görselleştirmek için kullanılır. Bu tür bir grafik, her hücreyi renkler aracılığıyla değerleri gösterir, genellikle en yüksek değerler daha koyu renkte olur. Örneğin, her eyalet için belirli bir dönemdeki maliyetler arasındaki korelasyonu görselleştirmek için kullanılabilir. Eğer iki değişken arasında güçlü bir ilişki varsa, bu renklerin yoğunluğuyla gösterilecektir.
# 8. Bar Chart (Çubuk Grafiği)
Çubuk grafiği, her bir eyaletin toplam gıda maliyetini karşılaştırmak için kullanılır. Eyaletlerin tüm verileri üzerinde sıralama yapılır ve her bir çubuğun uzunluğu, o eyaletin toplam maliyetine orantılıdır. Bu görselleştirme, hangi eyaletlerin daha yüksek maliyetlere sahip olduğunu net bir şekilde gösterir. Eyaletlerin sıralanması, en yüksekten en düşüğe doğru yapılır, bu da hangi eyaletlerin en fazla veya en az harcama yaptığını hızlıca görmenize olanak tanır.
